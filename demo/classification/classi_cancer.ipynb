TRAIN TEST SPLIT

```python
X = df.drop('benign_0__mal_1',axis=1).values
y = df['benign_0__mal_1'].values
```

```python
type(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```

SCALING THE DATA


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

CREATING THE MODEL

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape
```

```python
model=Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

TRAINING THE MODEL

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),verbose=1)
```

```python
model.history.history
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss
```

```python
model_loss.plot()
```

EARLY STOPPING

```python
model = Sequential()
model.add(Dense(units=30,activation='relu'))
model.add(Dense(units=15,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop=EarlyStopping(monitor='val_loss',mode='min',verbose=25,patience=25)
```

```python
model.fit(
    x=X_train, 
    y=y_train, 
    epochs=600,
    validation_data=(X_test, y_test), verbose=1,
    callbacks=[early_stop]
)
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

## Adding in DropOut Layers

```python
model = Sequential()
model.add(Dense(units=30,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=15,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=600,
          validation_data=(X_test, y_test), verbose=1,
          callbacks=[early_stop]
          )
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

## MODEL EVALUATION

```python
predictions = model.predict(X_test)
```

```python
if predictions.shape[1] == 1:  # If it's a single output (sigmoid)
    predicted_classes = (predictions > 0.5).astype("int32")
else:  # If it's a multi-class classification (softmax)
    predicted_classes = np.argmax(predictions, axis=1)

print(predicted_classes)
```

```python
predicted_classes
```

```python
y_test
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predicted_classes))
```

```python
print(confusion_matrix(y_test,predicted_classes))
```
